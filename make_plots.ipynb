```python
import ROOT
%jsroot
```

```python
f = ROOT.TFile('all_rec_hst.root')
f.ls()

hNumHitsReached = f.Get('hNumHitsReached')
hNumHitsMissed = f.Get('hNumHitsMissed')
hThetaMomTotal = f.Get('hThetaMomTotal')
hThetaMomReached = f.Get('hThetaMomReached')
hThetaMomMissed = f.Get('hThetaMomMissed')
hThetaMomAnnihil = f.Get('hThetaMomAnnihil')
```

```python
astack = ROOT.THStack('a',';Number of Tracker hits;')
hNumHitsMissed.SetLineColor(ROOT.kRed)

hNumHitsReached.SetTitle("Primary e+ hit the tracker")
hNumHitsMissed.SetTitle("Primary e+ did not hit the tracker")

astack.Add(hNumHitsReached)
astack.Add(hNumHitsMissed)

c = ROOT.TCanvas()
astack.Draw("nostack hist")
ROOT.gPad.BuildLegend(0.5,0.6,0.85,0.7)
# ROOT.gPad.SetLogy()
c.Draw()
```

```python
hTotal = hThetaMomTotal.ProjectionX()
hReached = hThetaMomReached.ProjectionX()

hThetaReached = ROOT.TGraphAsymmErrors(hReached,hTotal)
hThetaReached.SetTitle(";cos(theta);Primary e+ reached the tracker / total events")

c = ROOT.TCanvas()
hThetaReached.Draw()
c.Draw()
```

```python
hTotal = hThetaMomTotal.ProjectionY()
hReached = hThetaMomReached.ProjectionY()

hMomReached = ROOT.TGraphAsymmErrors(hReached,hTotal)
hMomReached.SetTitle(";Momentum (MeV);Primary e+ reached the tracker / total events")

c = ROOT.TCanvas()
hMomReached.Draw()
c.Draw()
```

```python
c = ROOT.TCanvas()
hThetaMomReached.Draw('colz')
c.Draw()

c1 = ROOT.TCanvas()
hTotal = hThetaMomTotal.ProjectionX()
hTotal.SetLineColor(ROOT.kRed)
hTotal.Draw('hist')
hThetaMomReached.ProjectionX().Draw('hist same')
c1.Draw()

c2 = ROOT.TCanvas()
hTotal = hThetaMomTotal.ProjectionY()
hTotal.SetLineColor(ROOT.kRed)
hTotal.Draw('hist')
hThetaMomReached.ProjectionY().Draw('hist same')
c2.Draw()
```

```python
hTotal = hThetaMomTotal.ProjectionX()
hReached = hThetaMomMissed.ProjectionX()

hMomMissed = ROOT.TGraphAsymmErrors(hReached,hTotal)

c = ROOT.TCanvas()
hMomMissed.Draw()
c.Draw()
```

```python
hTotal = hThetaMomTotal.ProjectionY()
hReached = hThetaMomMissed.ProjectionY()

hMomMissed = ROOT.TGraphAsymmErrors(hReached,hTotal)

c = ROOT.TCanvas()
hMomMissed.Draw()
c.Draw()
```

```python
c = ROOT.TCanvas()
hThetaMomMissed.Draw('colz')
c.Draw()

c1 = ROOT.TCanvas()
hThetaMomMissed.ProjectionX().Draw('hist')
c1.Draw()

c2 = ROOT.TCanvas()
hThetaMomMissed.ProjectionY().Draw('hist')
c2.Draw()
```
